 ### 2] Vaccination Rate

```python
url_2 = 'https://raw.githubusercontent.com/govex/COVID-19/master/data_tables/vaccine_data/global_data/time_series_covid19_vaccine_global.csv'
pd_raw_vaccine_data = pd.read_csv(url_2,sep=',')
```

```python
pd_raw_vaccine_data.head()
```

```python


df_compare = np.where(pd_raw_vaccine_data['Date'] == pd_raw_vaccine_data['Report_Date_String'], 'True', 'False')
print(np.where(df_compare == 'False'))

pd_raw_vaccine_data.dtypes
```

```python
# For Better Data Handling,  Dropping the first 'Date' column also dropping the "Province_State" columns. 

pd_raw_data=pd_raw_vaccine_data.drop(['Date', 'Province_State'], axis =1)


pd_raw_data = pd_raw_data.rename(columns = {'Country_Region':'Country', 'People_partially_vaccinated':'partially_vaccinated', 'People_fully_vaccinated':'fully_vaccinated', 'Report_Date_String':'Date'})

pd_raw_data.head()
```

```python
# Data type conversion to string.
pd_raw_data['Date'] = pd.to_datetime(pd_raw_data['Date'])
```

```python
pivot_data = pd.pivot_table(data = pd_raw_data,
                                  index = 'Date',
                                  columns = 'Country',
                                  values = 'fully_vaccinated',
                                  )

# Putting 0 for NAN values
pivot_data_fill_zero = pivot_data.fillna(0)
```

```python
pivot_data=pivot_data.replace(np.nan,0)

pivot_data.reset_index(inplace = True)

pivot_data.head()
```

```python
time_idx= pivot_data["Date"]
df_plot_values= pd.DataFrame({'Date': time_idx})
df_plot_values.head()
```

```python
country_list = ['Germany', 
                'US', 
                'Sweden']
```

```python
# Getting the data for the Global Population

url_pop = 'https://raw.githubusercontent.com/govex/COVID-19/master/data_tables/world_pop_by_country.csv'
pop = pd.read_csv(url_pop)
pop.head()
```

```python
for each in country_list:
    df_plot_values[each] = pivot_data[each]
    var = np.array(pop.loc[pop["Country Name"]==each, "2018"])
    df_plot_values[each] = df_plot_values[each]/var[0]*100 
    

    
df_plot_values.head()
```

```python
fig_1 = go.Figure()

for each in country_list:
    fig_1.add_trace(go.Scatter(x = df_plot_values.Date, 
                         y = df_plot_values[each], 
                        mode = 'markers+lines',
                        opacity = 0.9, 
                        line_width = 2,
                        marker_size = 2,
                        name = each))

fig_1.update_layout(width=800,
    height=600,
    title="The vaccination rate (percentage of the population) over time",
    title_font_color="Black",
    xaxis_title="Time Till Date",
    yaxis_title="Percentage Vaccination Rate",

    legend_title="Countries",
)

fig_1.update_yaxes(type = 'linear', range = [0, 90])




fig_1.show()
```
